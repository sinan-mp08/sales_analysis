-> changes shipdate and orderdate into datetime datatype YYYY-MM-DD

```python
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df.info()
```

```python
df.head()
```

-> convert money column to numbers and removed $

```python
money_cols = ['Sales', 'Profit', 'Cost','ProfitMargin','UnitPrice']

for col in money_cols:
    df[col] = df[col].replace(r'[$,]', '', regex=True).astype(float)
```

-> Fix Category Typos and Standardize Text Formatting

```python
obj_cols = df.select_dtypes(include=['object']).columns

for col in obj_cols:
    df[col] = (
        df[col]
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
        .str.title()
    )
```

-> Standardize the Returned Flag (Yes/No)

```python


df['Returned'] = (
    df['Returned']
    .astype(str)
    .str.strip()
    .str.lower()
    .replace({
        'y': 'Yes',
        'yes': 'Yes',
        'true': 'Yes',
        '1': 'Yes',
        'n': 'No',
        'no': 'No',
        'false': 'No',
        '0': 'No'
    })
)

# Verify the result
print(df['Returned'].unique())
```

-> Handling Missing Values and Invalid Quantity

```python

df.isnull().sum()
```

```python

df['PromoCode'] = df['PromoCode'].fillna('No Promo')
```

```python
df.isnull().sum()
```

```python
# Find Invalid Quantities (0 or Negative)

df[df['Quantity'] <= 0]
```

```python
df.to_csv('retail_sales_cleaned.csv', index=False)
```

### Monthly Sales Trend Analysis (Seasonality)

```python
import plotly.express as px
 
monthly_sales = (
    df.groupby("Month")["Sales"]
    .sum()
    .reset_index()
)

fig = px.line(
    monthly_sales,
    x="Month",
    y="Sales",
    markers=True,
    title="📈 Monthly Sales Trend",
    template="plotly_white"
)

fig.update_traces(
    line=dict(width=4),
    marker=dict(size=10)
)

fig.update_layout(
    title_x=0.5,
    hovermode="x unified",
    font=dict(size=16)
)

fig.show()
```

```python


region = df.groupby("Region").agg({
    "Sales":"sum",
    "Profit":"sum",
    "Quantity":"sum"
}).reset_index()

fig = px.scatter(
    region,
    x="Sales",
    y="Profit",
    size="Quantity",
    color="Region",
    hover_name="Region",
    size_max=60,
    template="plotly_dark",
    title="🌍 Sales & Profit by Region"
)

fig.update_layout(title_x=0.5)

fig.show()
```

### Sales by product category and sub-category

```python
import plotly.express as px

sales = (
    df.groupby(["Category", "SubCategory"])["Sales"]
      .sum()
      .reset_index()
)

fig = px.bar(
    sales,
    x="Sales",
    y="SubCategory",
    color="Category",
    orientation="h",
    text_auto=".2s",
    title="Sales by Product SubCategory",
    template="plotly_white"
)

fig.update_layout(
    title_x=0.5,
    height=700
)

fig.show()
```

## Year-over-year growth

```python
df["Year"] = df["OrderDate"].dt.year
yearly_sales = (
    df.groupby("Year")["Sales"]
      .sum()
      .reset_index()
)

print(yearly_sales)
```

```python
yearly_sales["YoY Growth (%)"] = (
    yearly_sales["Sales"].pct_change() * 100
)

print(yearly_sales)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Sales bars
fig.add_trace(go.Bar(
    x=yearly_sales["Year"],
    y=yearly_sales["Sales"],
    name="Sales"
))

# YoY Growth line
fig.add_trace(go.Scatter(
    x=yearly_sales["Year"],
    y=yearly_sales["YoY Growth (%)"],
    mode="lines+markers",
    name="YoY Growth %",
    yaxis="y2"
))

fig.update_layout(
    title="📊 Year-over-Year Growth Analysis",
    title_x=0.5,

    yaxis=dict(title="Sales"),

    yaxis2=dict(
        title="Growth %",
        overlaying="y",
        side="right"
    ),

    template="plotly_white",
    hovermode="x unified"
)

fig.show()
```

# Total Revenue & Profit Margin

```python
# Total Revenue
total_revenue = df["Sales"].sum()

# Total Profit
total_profit = df["Profit"].sum()

# Profit Margin (%)
profit_margin = (total_profit / total_revenue) * 100

print(f"Revenue: ₹{total_revenue:,.2f}")
print(f"Profit: ₹{total_profit:,.2f}")
print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

total_orders = df["OrderID"].nunique()

fig = make_subplots(
    rows=1,
    cols=4,
    specs=[[{"type":"indicator"}]*4]
)

fig.add_trace(go.Indicator(
    mode="number",
    value=total_revenue,
    number={'prefix':'$','valueformat':',.0f'},
    title={'text':'<b>Total Revenue</b>'}
), row=1, col=1)

fig.add_trace(go.Indicator(
    mode="number",
    value=total_profit,
    number={'prefix':'$','valueformat':',.0f'},
    title={'text':'<b>Total Profit</b>'}
), row=1, col=2)

fig.add_trace(go.Indicator(
    mode="number",
    value=profit_margin,
    number={'suffix':'%','valueformat':'.2f'},
    title={'text':'<b>Profit Margin</b>'}
), row=1, col=3)

fig.add_trace(go.Indicator(
    mode="number",
    value=total_orders,
    title={'text':'<b>Total Orders</b>'}
), row=1, col=4)

fig.update_layout(
    title="📈 Executive KPI Cards",
    title_x=0.5,
    template="plotly_white",
    height=300
)

fig.show()
```

# Average Order Value (AOV)

```python
# Total Revenue
total_revenue = df["Sales"].sum()

# Total Unique Orders
total_orders = df["OrderID"].nunique()

# Average Order Value
aov = total_revenue / total_orders

print(f"Average Order Value: {aov:,.2f}")
```

```python
fig = go.Figure(go.Indicator(
    mode="number",
    value=aov,
    number={
        "prefix": "$",
        "valueformat": ",.2f"
    },
    title={
        "text": "<b>🛒 Average Order Value (AOV)</b>"
    }
))

fig.update_layout(
    template="plotly_white",
    height=300,
    font=dict(size=18),
    margin=dict(t=60, b=30, l=30, r=30)
)

fig.show()
```

# Return rate & repeat-customer rate

```python


# Total unique orders
total_orders = df["OrderID"].nunique()

# Returned orders
returned_orders = df[df["Returned"] == "Yes"]["OrderID"].nunique()

# Return rate
return_rate = (returned_orders / total_orders) * 100

print(f"Return Rate: {return_rate:.2f}%")
```

```python
fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=return_rate,
    number={'suffix': '%'},
    title={'text': "📦 Return Rate"},
    gauge={
        'axis': {'range': [0, 100]},
        'bar': {'color': "crimson"},
        'steps': [
            {'range': [0, 10], 'color': '#d4edda'},
            {'range': [10, 20], 'color': '#fff3cd'},
            {'range': [20, 100], 'color': '#f8d7da'}
        ]
    }
))

fig.update_layout(
    template="plotly_white",
    height=400
)

fig.show()
```

```python
# Orders per customer
customer_orders = (
    df.groupby("CustomerID")["OrderID"]
      .nunique()
)

# Repeat customers
repeat_customers = (customer_orders > 1).sum()

# Total customers
total_customers = df["CustomerID"].nunique()

repeat_rate = (repeat_customers / total_customers) * 100

print(f"Repeat Customer Rate: {repeat_rate:.2f}%")
```

```python
repeat_df = pd.DataFrame({
    "Customer Type": ["Repeat", "New"],
    "Count": [
        repeat_customers,
        total_customers - repeat_customers
    ]
})

fig = px.pie(
    repeat_df,
    names="Customer Type",
    values="Count",
    hole=0.6,
    color="Customer Type",
    color_discrete_map={
        "Repeat": "#636EFA",
        "New": "#FFA15A"
    },
    title="👥 Repeat Customer Rate"
)

fig.update_layout(title_x=0.5)

fig.show()
```

```python


customer_funnel = pd.DataFrame({
    "Stage": [
        "Total Customers",
        "Repeat Customers"
    ],
    "Count": [
        total_customers,
        repeat_customers
    ]
})

fig = px.funnel(
    customer_funnel,
    x="Count",
    y="Stage",
    title="Customer Retention Funnel"
)

fig.update_layout(
    title_x=0.5,
    template="plotly_white"
)

fig.show()
```
